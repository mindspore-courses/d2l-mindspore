# 多层感知机

简要介绍一些常见的激活函数

```python
import sys
sys.path.append('..')
```

```python
%matplotlib inline
from d2l import mindspore as d2l
import mindspore as ms
import mindspore.ops as ops
```

ReLU提供了一种非常简单的非线性变换

```python
def relu(x):
    return ops.ReLU()(x)
```

```python
x = ops.arange(-8.0, 8.0, 0.1)
y = relu(x)
d2l.plot(x.asnumpy(), y.asnumpy(), 'x', 'relu(x)', figsize=(5, 2.5))
```

下面我们绘制ReLU函数的导数。

```python
grad_fn = ms.grad(relu, (0))
x_grad = grad_fn(x)
d2l.plot(x.asnumpy(), x_grad.asnumpy(), 'x', 'grad of relu', figsize=(5, 2.5))
```

对于一个定义域在$\mathbb{R}$中的输入，*sigmoid函数*将输入变换为区间(0, 1)上的输出

```python
def sigmoid(x):
    return ops.Sigmoid()(x)
```

```python
y = sigmoid(x)
d2l.plot(x.asnumpy(), y.asnumpy(), 'x', 'sigmoid(x)', figsize=(5, 2.5))
```

sigmoid函数的导数图像如下所示。

```python
grad_fn = ms.grad(sigmoid, (0))
x_grad = grad_fn(x)
d2l.plot(x.asnumpy(), x_grad.asnumpy(), 'x', 'grad of sigmoid', figsize=(5, 2.5))
```

Tanh(双曲正切)函数也能将其输入压缩转换到区间(-1, 1)上

```python
def tanh(x):
    return ops.Tanh()(x)
```

```python
y = tanh(x)
d2l.plot(x.asnumpy(), y.asnumpy(), 'x', 'tanh(x)', figsize=(5, 2.5))
```

tanh函数的导数图像如下所示。

```python
grad_fn = ms.grad(tanh, (0))
x_grad = grad_fn(x)
d2l.plot(x.asnumpy(), x_grad.asnumpy(), 'x', 'grad of tanh', figsize=(5, 2.5))
```
